```python
import requests
import pandas as pd
```

```python
headers = {
    'authoriy' : 'https://wiki.openstreetmap.org/'
}
response = requests.get('https://wiki.openstreetmap.org/wiki/List_of_London_Underground_stations', headers=headers)

# checking the status code and obtaining the html response content
if(response.status_code == 200):
    html_content = response.text
```

```python
all_tables = pd.read_html(html_content)
```

```python
underground_df = all_tables[0]

underground_df.columns.format()
```

```python
underground_df = underground_df.drop(['Platform / Entrance',
 'Collected By',
 'Collected On',
 'Line',
 'Step free'], axis=1)
```

```python
underground_df = underground_df.drop_duplicates(subset=['Name'], keep='first')
```
