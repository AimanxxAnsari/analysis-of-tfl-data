```python
import re
def clean_coordinates(df):
    def clean_coordinate(value):
        if isinstance(value, str):
            # Extract the main coordinate value before any +/- or parentheses
            match = re.search(r"^([-+]?\d*\.\d+|\d+)", value)
            if match:
                return float(match.group(1))
        elif isinstance(value, (float, int)):
            return value
        return None
    
    # Apply the cleaning function to Latitude and Longitude columns
    df['Latitude_cleaned'] = df['Latitude'].apply(clean_coordinate)
    df['Longitude_cleaned'] = df['Longitude'].apply(clean_coordinate)
    
    # Identify rows with invalid coordinates
    invalid_rows = df[df['Latitude_cleaned'].isna() | df['Longitude_cleaned'].isna()]
    
    # Remove rows with invalid coordinates from the main DataFrame
    cleaned_df = df.dropna(subset=['Latitude_cleaned', 'Longitude_cleaned'])
    
    # Drop the temporary cleaned columns used for validation
    cleaned_df = cleaned_df.drop(columns=['Latitude_cleaned', 'Longitude_cleaned'])
    invalid_rows = invalid_rows.drop(columns=['Latitude_cleaned', 'Longitude_cleaned'])
    
    return cleaned_df, invalid_rows
```

```python
tfl_df_c, tfl_df_invalid = clean_coordinates(tfl_df)
```
