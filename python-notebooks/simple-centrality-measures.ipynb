```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
## Degree Centrality

G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])
degree_centrality = nx.degree_centrality(G)

pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(5, 3))
nx.draw(G, pos, with_labels=False, node_color='lightblue', node_size=1000, edgecolors='black')
nx.draw_networkx_labels(G, pos, labels={node: f'{node}\n{cent:.2f}' for node, cent in degree_centrality.items()}, font_size=10, font_weight='bold')
plt.title("Degree Centrality", fontsize=16)
plt.show()
```

```python
## Betweenness Centrality

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (2, 5)])
betweenness_centrality = nx.betweenness_centrality(G)

pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(5, 3))
nx.draw(G, pos, with_labels=False, node_color='lightblue', node_size=1000, edgecolors='black')
nx.draw_networkx_labels(G, pos, labels={node: f'{node}\n{cent:.2f}' for node, cent in betweenness_centrality.items()}, font_size=10, font_weight='bold')
plt.title("Betweenness Centrality", fontsize=16)
plt.axis('off')
plt.show()
```

```python
## Closeness Centrality

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (1, 5)])
closeness_centrality = nx.closeness_centrality(G)

pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(5, 3))
nx.draw(G, pos, with_labels=False, node_color='lightblue', node_size=1000, edgecolors='black')
nx.draw_networkx_labels(G, pos, labels={node: f'{node}\n{cent:.2f}' for node, cent in closeness_centrality.items()}, font_size=10, font_weight='bold')
plt.title("Closeness Centrality", fontsize=16)
plt.axis('off')
plt.show()
```

```python
## Eigenvector Centrality

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3), (2, 4)])
eigenvector_centrality = nx.eigenvector_centrality(G)

pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(5, 3))
nx.draw(G, pos, with_labels=False, node_color='lightblue', node_size=1000, edgecolors='black')
nx.draw_networkx_labels(G, pos, labels={node: f'{node}\n{cent:.2f}' for node, cent in eigenvector_centrality.items()}, font_size=10, font_weight='bold')
plt.title("Eigenvector Centrality", fontsize=16)
plt.axis('off')
plt.show()
```

```python
## PageRank

G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3), (3, 1), (3, 2), (4, 2)])
pagerank = nx.pagerank(G)

pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(5, 3))
nx.draw(G, pos, with_labels=False, node_color='lightblue', node_size=1000, edgecolors='black', arrows=True, arrowsize=20)
nx.draw_networkx_labels(G, pos, labels={node: f'{node}\n{rank:.2f}' for node, rank in pagerank.items()}, font_size=10, font_weight='bold')
plt.title("PageRank", fontsize=16)
plt.axis('off')
plt.show()
```

```python
## Katz Centrality

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])
katz_centrality = nx.katz_centrality(G)

pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(5, 3))
nx.draw(G, pos, with_labels=False, node_color='lightblue', node_size=1000, edgecolors='black')
nx.draw_networkx_labels(G, pos, labels={node: f'{node}\n{cent:.2f}' for node, cent in katz_centrality.items()}, font_size=10, font_weight='bold')
plt.title("Katz Centrality", fontsize=16)
plt.axis('off')
plt.show()
```
